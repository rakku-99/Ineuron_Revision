```python
import math

n2=5
str="ineuron"
middle=math.floor(n2/2) if n2%2==0 else math.ceil(n2/2)
# first half
for l in range(1,n2+1):
    if l<=middle:
        k=middle-l
        for i in range(k):
            print(" "*len(str),end=" ")
        print((str+" "+" "*len(str)+" ")*l)
    else: # last half in reverse order
        r_spc=l-middle
        r= n2-l+1 #(5+1-4,5+1-5)
        for r2 in range(r):
            print(" "*len(str)*r_spc,str,end= " ")
        print("\n")
```
